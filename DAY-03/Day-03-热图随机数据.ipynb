```python
import pandas as pd
import os
import numpy as np
import json
import folium
import folium.plugins as plugins
```

```python
np.random.seed(3141592)
initial_data = (
    np.random.normal(size=(100, 2)) * np.array([[0.6, 0.6]]) +
    np.array([[34.27,108.95]])
)

move_data = np.random.normal(size=(100, 2)) * 0.01

# to_list() 将numpy数组转为list列表，使用列表推导式，会增加数据维度
data = [(initial_data + move_data * i).tolist() for i in range(100)]

weight = 1  # default value
for time_entry in data:
    for row in time_entry:
        row.append(weight)
```

```python
from datetime import datetime, timedelta
time_index = [
    (datetime.now()-timedelta(100) + k * timedelta(1)).strftime('%Y-%m-%d') for
    k in range(len(data))
]
```

```python
m = folium.Map([34.27,108.95], zoom_start=7)

hm = plugins.HeatMapWithTime(
    data,
    index=time_index,
    radius=20,
    auto_play=True,
    max_opacity=0.75,
    scale_radius=0
)

hm.add_to(m)

m
```

```python
m.save('heatMap.html')
```
