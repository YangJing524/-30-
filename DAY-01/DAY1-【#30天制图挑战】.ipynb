```python
import pandas as pd
import json
import time
import folium
import demjson
import requests
```

```python
data=pd.read_excel('医院.xlsx')
```

```python
data.head()
```

#### ADD Markers

```python
import random

list_range=tuple(random.randint(0,len(data)) for i in range(150))
                 
m = folium.Map(
    location=[34.27, 108.95],
    zoom_start=12,
    tiles='Stamen Toner'
)
for i in list_range:
    folium.CircleMarker(
    location=[data.iloc[i]['lat'], data.iloc[i]['long']],
    radius=8,
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(m)
    
    folium.Marker(
    location=[data.iloc[i]['lat'], data.iloc[i]['long']],
    popup=data.iloc[i]['name'],
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)
```
