```python
m = folium.Map(
    location=[45.372, -121.6972],
    zoom_start=12,
    tiles='Stamen Toner'
)

tooltip = '点击查看详情'

# folium.Marker([45.3288, -121.6625], popup='<i>Mt. Hood Meadows</i>', tooltip=tooltip).add_to(m)
folium.Marker([45.3311, -121.7113],
              popup=folium.map.Popup(
                  html='<b>Timberline Lodge open new line as frinebd</b></br><b>next line</b></br><b>next line</b></br>',
                  max_width=250), 
              tooltip=tooltip,
             draggable=True).add_to(m)

m
```

```python
m = folium.Map(
    location=[45.372, -121.6972],
    zoom_start=12,
    tiles='Stamen Terrain'
)
folium.Marker([45.372, -121.6972], 
    icon=plugins.BeautifyIcon(icon='loding',
    iconShape= 'marker',
    background_color='red',
    border_color='blue',
    number=5,
    ICON_SHAPE_TYPES='circle',
    text_color='blue'),
    draggable=True,
    popup='popup',
    tooltip='This is a beatifyMarker'
).add_to(m)

m
```
